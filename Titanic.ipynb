```python
#Kaç yolcu hayatta kaldı, kaç yolcu hayatını kaybetti?
import pandas as pd
import matplotlib.pyplot as plt

titanic_verisi = pd.read_csv('titanic.csv')
hayatta_kalma = titanic_verisi['Survived'].value_counts()


plt.figure(figsize=(10,5))
plt.bar(['Hayatta Kalan', 'Hayatta Kalamayan'], hayatta_kalma, color=['green', 'red'])
plt.xlabel('Yolcu durumu')
plt.ylabel('Yolcu Sayısı')
plt.title('Hayatta Kalan ve Hayatını Kaybeden Yolcuların Dağılımı')

plt.show()


```

```python
#Hayatta kalanların cinsiyet dağılımı nasıldı?
import pandas as pd
import matplotlib.pyplot as plt

titanic_verisi = pd.read_csv('titanic.csv')
cinsiyet_dagilimi = titanic_verisi['Sex'].value_counts()
plt.pie(cinsiyet_dagilimi, labels=cinsiyet_dagilimi.index, autopct='%1.1f%%', colors=['skyblue', 'pink'])
plt.title('Cinsiyete Göre Hayatta Kalma Durumu')

plt.show()
```

```python
#Hayatta kalanların yaş dağılımı nasıldı?
import pandas as pd
import matplotlib.pyplot as plt

titanic_verisi = pd.read_csv('titanic.csv')

hayatta_kalanlar = titanic_verisi[titanic_verisi['Survived'] == 1]
plt.figure(figsize=(10, 6))
plt.hist(hayatta_kalanlar['Age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Yaş')
plt.ylabel('Yolcu Sayısı')
plt.title('Hayatta Kalan Yolcuların Yaş Dağılımı')


plt.show()

```
