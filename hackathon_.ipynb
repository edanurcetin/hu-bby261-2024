```python
# 1)Kullanıcıdan gerekli verileri alacağız: başlık, konu, açıklama, oluşturucu, lisans, dil, URL.
# 2) Bu verileri XML formatında saklayacağız.
# 3)Kaydedilen XML dosyasına yazacağız.

import xml.etree.ElementTree as ET

root = ET.Element("OERMetadata")

def veri_topla():
    title = input("Kaynağın Başlığını giriniz: ")
    subject = input("Anahtar Kelimeleri giriniz: ")
    description = input("Açıklamayı giriniz: ")
    creator = input("Oluşturan kişiyi / kuruluşu giriniz: ")
    license = input("Lisans bilgisini giriniz: ")
    language = input("Dili giriniz: ")
    identifier = input("Kaynağın URL'sini giriniz: ")


    kaynak = ET.Element("Kaynak")
    ET.SubElement(kaynak, "Title").text = title
    ET.SubElement(kaynak, "Subject").text = subject
    ET.SubElement(kaynak, "Description").text = description
    ET.SubElement(kaynak, "Creator").text = creator
    ET.SubElement(kaynak, "License").text = license
    ET.SubElement(kaynak, "Language").text = language
    ET.SubElement(kaynak, "Identifier").text = identifier


    root.append(kaynak)


def veri_kaydet():
    veri_topla()


    tree = ET.ElementTree(root)
    tree.write("oer_metadata.xml", encoding="utf-8", xml_declaration=True)
    print("Veriler başarıyla kaydedildi.")

veri_kaydet()
```

```python
# XML dosyasından verileri okuma.
# Kullanıcının istediği bir arama işlemi yapabilmesi.
# Kaynakları Listelemesi

import xml.etree.ElementTree as ET


def veri_okuma():

    tree = ET.parse("oer_metadata.xml")
    root = tree.getroot()


    print("Kaynaklar:")
    for kaynak in root.findall("Kaynak"):
        title = kaynak.find("Title").text
        subject = kaynak.find("Subject").text
        description = kaynak.find("Description").text
        creator = kaynak.find("Creator").text
        license = kaynak.find("License").text
        language = kaynak.find("Language").text
        identifier = kaynak.find("Identifier").text


        print(f"Başlık: {title}")
        print(f"Konu: {subject}")
        print(f"Açıklama: {description}")
        print(f"Oluşturan: {creator}")
        print(f"Lisans: {license}")
        print(f"Dil: {language}")
        print(f"URL: {identifier}")
        print("-" * 40)


def arama_yap():
    arama_kelimesi = input("Arama yapmak istediğiniz kelimeyi girin: ").lower()


    tree = ET.parse("oer_metadata.xml")
    root = tree.getroot()

    print("Arama Sonuçları:")
    for kaynak in root.findall("Kaynak"):
        title = kaynak.find("Title").text.lower()
        subject = kaynak.find("Subject").text.lower()
        description = kaynak.find("Description").text.lower()


        if arama_kelimesi in title or arama_kelimesi in subject or arama_kelimesi in description:
            print(f"Başlık: {kaynak.find('Title').text}")
            print(f"Konu: {kaynak.find('Subject').text}")
            print(f"Açıklama: {kaynak.find('Description').text}")
            print("-" * 40)


def menu():
    while True:
        print("\nSeçenekler:")
        print("1. Kaynakları listele")
        print("2. Arama yap")
        print("3. Çıkış")

        secim = input("Seçiminizi yapın (1/2/3): ")

        if secim == "1":
            veri_okuma()
        elif secim == "2":
            arama_yap()
        elif secim == "3":
            print("Çıkılıyor...")
            break
        else:
            print("Geçersiz seçim. Lütfen tekrar deneyin.")


menu()
```
